```python
# importing pandas module
import pandas as pd

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# reading housing data
housing_data = pd.read_excel('25_SpringHouses.xlsx.')

# showing sample data
housing_data.head()

# describing data
housing_data.info()

# converting data types
# housing_data["Baths"] = housing_data["Baths"].astype("str")
# housing_data["Beds"] = housing_data["Beds"].astype("str")
housing_data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
# create boxplot using seaborn
sns.boxplot(y="Selling Price", data=housing_data,color="brown")
plt.title('House Sell Price Boxplot Distribution')

# show the plot
plt.show()

plt.figure(figsize=(12,8))
# create boxplot using seaborn
sns.histplot(data=housing_data,x="Selling Price",color="brown")
plt.title('House Sell Price Histogram Distribution')

# show the plot
plt.show()
```

```python
plt.figure(figsize=(12,8))
# create boxplot using seaborn
sns.boxplot(y="Sq Ft", data=housing_data,color="green")
plt.title('House Sq Ft Boxplot Distribution')

# show the plot
plt.show()

housing_data.describe()

housing_data.describe(
    percentiles = [.05,.25, .5, .75, .95]).apply(lambda s: s.apply('{0:,.8f}'.format))

plt.figure(figsize=(12,8))
# create boxplot using seaborn
sns.histplot(data=housing_data,x="Sq Ft",color="green")
plt.title('House Sq Ft Histogram Distribution')

# show the plot
plt.show()
```

```python
# creating temporary data frames to investigate outliers records for fertiliser column where value is more than 95% percentile
temp_housing_data_out = housing_data[housing_data["Sq Ft"]>3424.00]

# sorting records based on high fertiliser amount in descending  order
temp_housing_data_out.sort_values(by = ["Sq Ft"], ascending = False)[:10]
```

```python
# calculating correlation
corrmat = housing_data.corr()
corrmat

# heatmap of correlation matrix
f, ax = plt.subplots(figsize = (9, 6))
sns.heatmap(corrmat, vmax = .8, square = True, annot = True, cmap = 'RdYlBu', linewidths = .5 )
```

```python
sns.pairplot(housing_data)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Beds",y="Selling Price",data=housing_data)
plt.title('House Sell Price Boxplot Distribution for Different Beds numbers')
plt.show()

plt.figure(figsize=(12,8))
# create boxplot using seaborn
sns.histplot(data=housing_data,x="Beds")
plt.title('House Beds Histogram Distribution')

# show the plot
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="Sq Ft",y="Selling Price",data=housing_data,color = "purple")
plt.title('House Sell Price based on Sq Ft area')
plt.show()

plt.figure(figsize=(12,8))
sns.regplot(x="Sq Ft",y="Selling Price",data=housing_data,color = "purple")
plt.title('House Sell Price based on Sq Ft area along with Least Square line')
plt.show()
```
