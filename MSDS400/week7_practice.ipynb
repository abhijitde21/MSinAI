```python
import numpy as np
import sympy as sym

def f(x):
    return x**2 + 4 

x = sym.Symbol('x')
int1 = sym.integrate(f(x),(x, 0, 2))
print(int1)
```

```python
def f(x):
    return sym.exp(x)

x = sym.Symbol('x')
int1 = sym.integrate(f(x),(x, 0, 2))
print(int1)
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

mu = 2
sg = 1

x_range = np.linspace(-10, 10, 100)
norm_pdf = norm.pdf(x_range, mu, sg)
plt.plot(x_range,norm_pdf)
plt.xlabel('x-data')
plt.ylabel('pdf_value')
plt.title("PDF of a Normal Distribution with mean=2 and sigma=1")
plt.show()
```

```python
norm_cdf = norm.pdf(x_range, mu, sg)
plt.plot(x_range,norm_cdf)
plt.xlabel('x-data')
plt.ylabel('cdf_value')
plt.title("CDF of a Normal Distribution with mean=2 and sigma=1")
plt.show()
```
