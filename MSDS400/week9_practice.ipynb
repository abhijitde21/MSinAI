```python
import networkx as nx
import matplotlib.pyplot as plt

# list of routes
edgelist = [['Mumbai', 'Kolkata', 2200], ['Mumbai', 'Delhi', 1500], ['Mumbai', 'Chennai', 3000], ['Mumbai', 'Indore', 700], 
            ['Kolkata', 'Delhi', 1600], ['Kolkata', 'Chennai', 3200], ['Kolkata', 'Indore', 1000], 
            ['Delhi', 'Chennai', 6000], ['Delhi', 'Indore', 2800], ['Chennai', 'Indore', 2700]]

# adding edges
g = nx.Graph()
for edge in edgelist:
    g.add_edge(edge[0],edge[1], weight = edge[2])   
    
# position of the graph
pos=nx.spring_layout(g)

# nodes
nx.draw_networkx_nodes(g, pos, node_size=600)

# edges
nx.draw_networkx_edges(g, pos, width=2)
nx.draw_networkx_edges(g, pos, width=2,  edge_color="g", style="dashed")

# node labels
nx.draw_networkx_labels(g, pos, font_size=10)

# edge weight labels
edge_labels = nx.get_edge_attributes(g, "weight")
nx.draw_networkx_edge_labels(g, pos, edge_labels)

plt.show()
```

```python
# calculating soanning tree
spanning_tree = nx.minimum_spanning_tree(g)
print(spanning_tree)
nx.draw_networkx(spanning_tree)
```

```python
# calculating shortest path from Mumbai to Kolkata using Dijkstra Algorithm
dijpath = nx.dijkstra_path(g, source='Mumbai', target='Kolkata')
print(dijpath)
```
